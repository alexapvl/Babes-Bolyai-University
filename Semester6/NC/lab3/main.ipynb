```python
import pandas as pd


def divided_differences_table(nodes, f=None, values=None):
    n = len(nodes)
    table = [[0 for _ in range(n + 1)] for _ in range(n)]

    if values is not None and len(values) != n:
        raise ValueError("'values' must have the same length as 'nodes'.")

    if f is None and values is None:
        raise ValueError("Provide either a function 'f' or a list 'values'.")

    # Column 0: x values
    for i in range(n):
        table[i][0] = nodes[i]

    # Column 1: f(x)
    for i in range(n):
        if values is not None:
            table[i][1] = values[i]
        else:
            table[i][1] = f(nodes[i])

    # Columns 2..n: divided differences
    # table[i][j] corresponds to f[x_i, ..., x_{i + j - 1}]
    for j in range(2, n + 1):
        for i in range(0, n - j + 1):
            numerator = table[i + 1][j - 1] - table[i][j - 1]
            denominator = nodes[i + j - 1] - nodes[i]
            table[i][j] = numerator / denominator

    columns = ["x", "f(x)"] + [f"order_{k}" for k in range(1, n)]
    return pd.DataFrame(table, columns=columns)


nodes = [0, 1, 2]
f = lambda x: 1 / (1 + x)

print("Exercise 1 a)")
dd_table = divided_differences_table(nodes, f)
dd_table
```

```python
import pandas as pd


def divided_differences_double_nodes(nodes, f, df):
    """
    Divided differences table for double nodes (Hermite interpolation).
    Each node x_i is repeated twice.

    Parameters:
        nodes: [x0, x1, ..., xn]
        f:    function f(x)
        df:   derivative f'(x)
    """
    n = len(nodes)
    m = 2 * n  # number of rows after doubling nodes

    z = [0.0] * m
    table = [[0 for _ in range(m + 1)] for _ in range(m)]

    # Build doubled nodes and first two columns
    for i in range(n):
        z[2 * i] = nodes[i]
        z[2 * i + 1] = nodes[i]

        table[2 * i][0] = z[2 * i]
        table[2 * i + 1][0] = z[2 * i + 1]

        table[2 * i][1] = f(nodes[i])
        table[2 * i + 1][1] = f(nodes[i])

    # First-order divided differences:
    # - use derivative only for division-by-zero case (repeated node)
    # - otherwise use the standard formula
    for i in range(m - 1):
        denominator = z[i + 1] - z[i]
        if denominator == 0:
            table[i][2] = df(z[i])
        else:
            table[i][2] = (table[i + 1][1] - table[i][1]) / denominator

    # Higher-order divided differences
    for j in range(3, m + 1):
        for i in range(0, m - j + 1):
            numerator = table[i + 1][j - 1] - table[i][j - 1]
            denominator = z[i + j - 1] - z[i]
            table[i][j] = numerator / denominator

    columns = ["z", "f(z)"] + [f"order_{k}" for k in range(1, m)]
    return pd.DataFrame(table, columns=columns)


# Example usage:
nodes = [0, 1, 2]
f = lambda x: 1 / (1 + x)
df = lambda x: -1 / (1 + x)**2

print("Exercise 1 b)")
hermite_table = divided_differences_double_nodes(nodes, f, df)
hermite_table
```

```python
import pandas as pd


def finite_differences_table(nodes, values, mode="forward"):
    """
    Build finite differences table from tabulated values.

    mode:
        - "forward" (default)
        - "backward"
        - "simple" (alias of forward)
    """
    if len(nodes) != len(values):
        raise ValueError("'nodes' and 'values' must have the same length.")

    mode = mode.lower()
    if mode == "simple":
        mode = "forward"
    if mode not in {"forward", "backward"}:
        raise ValueError("mode must be 'forward', 'backward', or 'simple'.")

    n = len(values)
    table = [[0 for _ in range(n + 1)] for _ in range(n)]

    # Base columns
    for i in range(n):
        table[i][0] = nodes[i]
        table[i][1] = values[i]

    prev = values[:]
    for order in range(1, n):
        if mode == "forward":
            curr = [prev[i + 1] - prev[i] for i in range(len(prev) - 1)]
        else:  # backward
            curr = [prev[i] - prev[i - 1] for i in range(1, len(prev))]

        for i, val in enumerate(curr):
            if mode == "forward":
                # Forward differences are aligned at the top.
                row = i
            else:
                # Backward differences are aligned at the bottom.
                row = i + order
            table[row][order + 1] = val

        prev = curr

    columns = ["x", "f(x)"] + [f"delta_{k}" for k in range(1, n)]
    return pd.DataFrame(table, columns=columns)


print("Exercise 2 b) - Forward differences")
forward_table_ex2 = finite_differences_table(nodes_ex2, values_ex2, mode="forward")
forward_table_ex2
```

```python
# Exercise 2 c)
# Backward differences
backward_table_ex2 = finite_differences_table(nodes_ex2, values_ex2, mode="backward")
backward_table_ex2
```
