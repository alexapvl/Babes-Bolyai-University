# Exercise 1
- Graph on the same set of axes the function e^x and its Taylor polynomials of degrees 1, 2, 3 and 4, for x ∈ [−3, 3];
- Approximate e with 6 correct decimals.

```python
import numpy as np
import matplotlib.pyplot as plt

# x range
x = np.linspace(-3, 3, 400)

# true function
y = np.exp(x)

# Taylor polynomials at 0 (Maclaurin)
P1 = 1 + x
P2 = 1 + x + x**2 / 2
P3 = 1 + x + x**2 / 2 + x**3 / 6
P4 = 1 + x + x**2 / 2 + x**3 / 6 + x**4 / 24

plt.figure(figsize=(8, 5))
plt.plot(x, y, 'k', label='e^x')
plt.plot(x, P1, label='P1(x)')
plt.plot(x, P2, label='P2(x)')
plt.plot(x, P3, label='P3(x)')
plt.plot(x, P4, label='P4(x)')

plt.xlabel('x')
plt.ylabel('y')
plt.title('e^x and its Maclaurin polynomials (degrees 1–4)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import math

target = 0.5e-6  # 10^{-6} / 2
s = 0.0
k = 0
term = 1.0

while term > target:
    s += term
    k += 1
    term = 1 / math.factorial(k)

print("Approximation of e:", s)
print("Number of terms used:", k)
print("math.e for comparison:", math.e)
print("Error:", abs(math.e - s))
```

# Exercise 2
 - Graph on the same set of axes the function sin x and its Taylor polynomials of degrees 3 and 5, for x ∈ [−π, π];
 - Approximate sin(π/5) with 5 correct decimals;

```python
import numpy as np
import matplotlib.pyplot as plt

# x range
x = np.linspace(-np.pi, np.pi, 400)

# true function
y = np.sin(x)

# Taylor polynomials at 0 (Maclaurin)
P3 = x - x**3 / 6
P5 = x - x**3 / 6 + x**5 / 120

plt.figure(figsize=(8, 5))
plt.plot(x, y, 'k', label='sin(x)')
plt.plot(x, P3, label='P3(x)')
plt.plot(x, P5, label='P5(x)')

plt.xlabel('x')
plt.ylabel('y')
plt.title('sin(x) and its Maclaurin polynomials (degrees 3 and 5)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import math

x = math.pi / 5     # point where we approximate
target = 0.5e-5     # 10^{-5} / 2

s = 0.0
k = 0
term = x           # first term of sin series
sign = 1

while abs(term) > target:
    s += term
    k += 1
    # next term: (-1)^k * x^(2k+1) / (2k+1)!
    term = ((-1)**k) * x**(2*k+1) / math.factorial(2*k+1)

print("Number of terms: ", k)
print("Approximation of sin(pi/5):", s)
print("math.sin(pi/5) for comparison:", math.sin(x))
print("Error:", abs(math.sin(x) - s))
```

```python
import math

print("=== Exercise 2 b) ===")
print("We use the alternating series: ln 2 = 1 - 1/2 + 1/3 - 1/4 + ...")

target_error = 0.5e-5  # 10^{-5} / 2

# For an alternating series with decreasing terms, the error after n terms
# is at most the size of the next term: 1/(n+1)
n = math.ceil(1 / target_error) - 1    # solve 1/(n+1) < target_error

print(f"Target error: {target_error:.1e}")
print(f"Condition: 1/(n+1) < {target_error:.1e}")
print(f"Smallest n satisfying this: n = {n}")

# (Optional) compute partial sum with n terms
s = 0.0
for k in range(1, n + 1):
    s += ((-1)**(k+1)) / k

true_val = math.log(2)
err = abs(true_val - s)

print(f"\nApproximation of ln(2) with {n} terms: {s:.10f}")
print(f"True ln(2):                          {true_val:.10f}")
print(f"Absolute error:                       {err:.3e}")
```

```python
import math
import numpy as np
import matplotlib.pyplot as plt

print("=== Exercise 2 c) ===")
print("Series for ln(1 - x):")
print("ln(1 - x) = - (x + x^2/2 + x^3/3 + x^4/4 + ...) ")
print("           = - sum_{k=1}^∞ x^k / k   for |x| < 1\n")

def ln1_minus_x_series_scalar(x, n_terms):
    """Scalar version with prints for a single x."""
    s = 0.0
    print(f"Approximating ln(1 - x) at x = {x}")
    for k in range(1, n_terms + 1):
        term = -x**k / k
        s += term
        if k <= 5 or k == n_terms:  # don't spam: show first few and last
            print(f"  term {k}: {term:.10f}   partial sum: {s:.10f}")
    return s

# Example: approximate ln(1 - 0.5)
x_example = 0.5
n_terms_example = 10
approx_example = ln1_minus_x_series_scalar(x_example, n_terms_example)
true_example = math.log(1 - x_example)

print(f"\nSeries approximation (n={n_terms_example}) at x = {x_example}: {approx_example:.10f}")
print(f"True value ln(1 - {x_example}):                {true_example:.10f}")
print(f"Absolute error:                                {abs(true_example - approx_example):.3e}\n")

# Vectorized version for plotting
def ln1_minus_x_series(x, n_terms):
    s = 0.0
    for k in range(1, n_terms + 1):
        s -= x**k / k
    return s

x_vals = np.linspace(-0.9, 0.9, 400)
n_terms_plot = 10

true_vals = np.log(1 - x_vals)
series_vals = ln1_minus_x_series(x_vals, n_terms_plot)

plt.figure(figsize=(8, 5))
plt.plot(x_vals, true_vals, 'k', label='ln(1 - x)')
plt.plot(x_vals, series_vals, label=f'Series (n={n_terms_plot})')
plt.xlabel('x')
plt.ylabel('y')
plt.title('ln(1 - x) vs Taylor series approximation')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import math
import numpy as np
import matplotlib.pyplot as plt

print("=== Exercise 2 d) ===")
print("Series for ln((1 + x)/(1 - x)):")
print("ln((1 + x)/(1 - x)) = 2 * (x + x^3/3 + x^5/5 + x^7/7 + ...)")
print("                     = 2 * sum_{m=0}^∞ x^{2m+1} / (2m+1)   for |x| < 1\n")

def ln_fraction_series_scalar(x, n_terms):
    """Scalar version with prints for a single x."""
    s = 0.0
    print(f"Approximating ln((1 + x)/(1 - x)) at x = {x}")
    for m in range(n_terms):
        power = 2*m + 1
        term = x**power / power
        s += term
        if m < 4 or m == n_terms - 1:  # show first few and last
            print(f"  m={m}, power={power}: term={term:.10f}, partial sum={s:.10f}")
    result = 2 * s
    print(f"After multiplying by 2: {result:.10f}")
    return result

# Example: approximate at x = 0.5
x_example = 0.5
n_terms_example = 10
approx_example = ln_fraction_series_scalar(x_example, n_terms_example)
true_example = math.log((1 + x_example) / (1 - x_example))

print(f"\nSeries approximation (n={n_terms_example}) at x = {x_example}: {approx_example:.10f}")
print(f"True value ln((1 + {x_example})/(1 - {x_example})): {true_example:.10f}")
print(f"Absolute error:                                   {abs(true_example - approx_example):.3e}\n")

# Vectorized version for plotting
def ln_fraction_series(x, n_terms):
    s = 0.0
    for m in range(n_terms):
        power = 2*m + 1
        s += x**power / power
    return 2 * s

x_vals = np.linspace(-0.9, 0.9, 400)
n_terms_plot = 10

true_vals = np.log((1 + x_vals) / (1 - x_vals))
series_vals = ln_fraction_series(x_vals, n_terms_plot)

plt.figure(figsize=(8, 5))
plt.plot(x_vals, true_vals, 'k', label='ln((1 + x)/(1 - x))')
plt.plot(x_vals, series_vals, label=f'Series (n={n_terms_plot})')
plt.xlabel('x')
plt.ylabel('y')
plt.title('ln((1 + x)/(1 - x)) vs Taylor series approximation')
plt.grid(True)
plt.legend()
plt.show()
```
